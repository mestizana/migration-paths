```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import imageio
import os
import plotly.io as pio
from plotly.subplots import make_subplots
import numpy as np
import plotly.graph_objects as go
import geopandas as gpd
```

```python
df = pd.read_csv('output.csv')
df
```

```python
df2 = df.drop(['Refugee Status Type', 'Area' , 'Category' , 'Recognized Cases' , 'Rejected Cases' , 'Pending Cases' , 'Total Applications' , 'Female 18-59', 'Female 60+', 'F: Unknown' , 
                'Male 18-59' , 'Male 60+' , 'M: Unknown', 'Total Females' , 'Total Males'], axis=1)
```

```python
df2.dtypes
```

```python
df2.drop_duplicates(inplace = True)
```

```python
df2
```

```python
filtered_df = df2[df2['Year'].isin([2013, 2014, 2015])]
```

```python
filtered_df['Sum of Total Children'] = filtered_df['Female 0-4'] + filtered_df['Female 5-17'] + filtered_df['Male 0-4'] + filtered_df['Male 5-17']
```

```python
fig = px.scatter(filtered_df, x='Year', y='Sum of Total Children', trendline='ols')

fig.update_layout(
    width=800,
    height=500,
    title='Sum of Total Children Migrating (2013-2015)',
    xaxis_title='Year',
    yaxis_title='Sum of Total Children'
)

fig.show(renderer = 'browser')
```
